```python
# input
# N = banyaknya jarum
# l = panjang jarum. tidak boleh lebih dari d
# d = lebar garis
# Q = total jarum yang menyentuh garis
# 0 <= x <= d dimana x = l*cos(theta)
# 0 <= theta <= 90

# Hitung nilai W = (2*l*N)/(d*Q)
```

```python
import random 
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
l = 4.5
d = 6
loop = 10
ex = []
ye = []
# banyaknya jarum dari 100 hingga 10.100
N = 100
while N <= 10100:
    ptot = 0
    for i in range(loop):
        Q = 0
        for j in range(N):
            theta = random.uniform(0,math.pi/2)
            x = random.uniform(0,d)
            if x <= l*math.cos(theta):
                Q = Q + 1
        p = (2*l*N)/(d*Q)
        ptot = ptot + p
    ex.append(N)
    ye.append(ptot/loop)
    N = N + 100
```

```python
plt.plot(ex,ye)
plt.show()
```
